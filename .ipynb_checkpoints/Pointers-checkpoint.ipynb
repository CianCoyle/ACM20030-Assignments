```python
def f(x):
    return np.sin(x)

def g(x):
    return -np.sin(x)
```

```python
x = np.arange(-2,8,.01)
vf = np.vectorize(f)
y = vf(x)
vg = np.vectorize(g)
h = vg(x)

plt.plot(x,h, "mo")
plt.plot(x,y, "co")
plt.title("Graph of Sin(X) & -Sin(X)")
plt.xlabel("X Value")
plt.ylabel("Output")

# m is pink
# c is turquoise
# k is black
plt.grid(True)
```
