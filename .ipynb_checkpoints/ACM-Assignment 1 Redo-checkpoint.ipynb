```python
# Q 1a
def VectorLength(x,y):
    answer = np.sqrt(x**2 + y**2)
    return answer
```

```python
# Q 2a
def IsPrime(n):
    i = 2
    while i < np.sqrt(n):
        if(n%i==0):
            return 0
        i += 1
    return 1

IsPrime(1), IsPrime(2), IsPrime(3), IsPrime(4), IsPrime(5), IsPrime(6), IsPrime(7), IsPrime(8), IsPrime(9), IsPrime(10)
# The function returns 1 and 4 & 9 (as well as the squares of all primes) as being prime ) as primes
```

```python
# Q 2c
def IsPrime(n):
    if n==1:
        return 0
# This fixes the problem that 1 was being stated to be prime which is false
    i = 2
    while i <= np.sqrt(n):
# In the previous line, i added the equals sign so that the squares of primes are not returned as being prime
        if(n%i==0):
            return 0
        i += 1
    return 1

IsPrime(1), IsPrime(9), IsPrime(49) 
```

```python
# Q 2e
def PrimeList(n):
    answer = list()
    for x in range(1,n+1):
        if (IsPrime(x)==1):
            answer.append(x)
    return answer
```

```python
PrimeList(1000)
```

```python
# Q 2f
def PrimeCount(n):
    answer = len(PrimeList(n))
    return answer
```

```python
# Q 2f continued
PrimeCount(1000)
```

```python
# Q 2g
x = np.arange(0,40,1)
Pi = np.zeros(40)
i = 0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i += 1

plt.plot(x, Pi, 'ro')
plt.xlabel('X')
plt.ylabel('Pi(X)')
plt.grid(True)
plt.title('Graph of X vs. Pi(X)')
```
