```python
def VectorLength(x,y):
    result = np.sqrt(x**2 + y**2)
    return result
```

```python
print("Answer to Q1-A")
print("Length of vector (5,6) is", VectorLength(5,6))
```

```python
print("Answer to Q1-C")
# Function to check if a number is prime
def IsPrime(n):
    if(n == 1):
        return 0
# The previous two lines solve the issue that one is not prime
    i = 2
    while i <= np.sqrt(n):
    # For each i, check if it divides n
    # Added a = sign after the < as this means it is less than or equal to
        if(n % i == 0):
            return 0
        i += 1
    return 1
# If no divisors are found, the number is prime
IsPrime(9)
```

```python
print("Answer to Q1-E")
def PrimesList(n):
    result = list()
    for x in range(1, n+1):
        prime = True
        for i in range(2, x):
            if (x % i == 0):
                prime = False
        if prime:
            result.append(x)
    return result
PrimesList(1000)
```

```python
print("Answer to Q1-F")
def PrimeCount(n):
    result = len(PrimesList(n))
    return result

print("PrimeCount(1000) =", PrimeCount(1000))
```

```python
print("Answer to Q1-G")
x = np.arange(0,40,1)
Pi = np.zeros(40)
i = 0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i += 1
plt.plot(x, Pi, "ro")
plt.xlabel('Input')
plt.ylabel('yLabel')
```
