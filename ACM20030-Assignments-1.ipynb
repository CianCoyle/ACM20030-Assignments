```python
def VectorLength(x,y):
    result = np.sqrt(x**2 + y**2)
    return result
```

```python
print("Answer to Q1-C")
# Function to check if a number is prime
def IsPrime(n):
    if(n == 1):
        return 0
# The previous two lines solve the issue that 1 is not prime
    i = 2
    while i <= np.sqrt(n):
        # For each i, check if it divides n
        # Added = sign after the < as this means less than or equal to
        if(n % i == 0):
            return 0
        i += 1
    return 1
# If no divisors are found, the number is prime
```

```python
def PrimeList(n):
    answer = list(1)
    for x in range(1,n+1):
        if (IsPrime(x)==1):
            answer.append(x)
    return answer
# n is the number up to which the primes are listed
```

```python
print("Answer to Q1-E")
PrimeList(1000)
```

```python
def PrimeCount(n):
    result = len((PrimeList(n)))
    return result
```

```python
print("Answer to Q1-F")
print("There are", PrimeCount(1000), "prime numbers between 1 and 1000")
```

```python
print("Answer to Q1-G")
x = np.arange(0,40,1)
Pi = np.zeros(40)
i = 0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i += 1
plt.plot(x, Pi, "ro")
plt.xlabel ('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
```
