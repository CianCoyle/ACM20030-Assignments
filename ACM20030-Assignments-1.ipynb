```python
# Q2a-c
# Function to check if a number is prime
def IsPrime(n):
    if(n == 1):
        return 0
# The previous two lines solve the issue that 1 is not prime
    i = 2
    while i <= np.sqrt(n):
        # For each i, check if it divides n
        # Added = sign after the < as this means less than or equal to
        if(n % i == 0):
            return 0
        i += 1
    return 1
# If no divisors are found, the number is prime
```

```python
# Q2e
def PrimeList(n):
    answer = list()
    for x in range(1,n+1):
        if (IsPrime(x)==1):
            answer.append(x)
    return answer
# n is the number up to which the primes are listed
```

```python
# Q2e
PrimeList(1000)
```

```python
# Q2f
def PrimeCount(n):
    result = len((PrimeList(n)))
    return result
```

```python
# Q2f
print("There are", PrimeCount(1000), "prime numbers between 1 and 1000")
```

```python
# Q2g
x = np.arange(0,40,1)
Pi = np.zeros(40)
i = 0
while i < np.size(Pi):
    Pi[i] = PrimeCount(i)
    i += 1
plt.plot(x, Pi, "ro")
plt.xlabel ('X')
plt.ylabel('Pi(x)')
plt.grid(True)
```

```python
# Q3b
data_1 = np.loadtxt("PowerLawOrExponential.txt")
```

```python
# Q3b
X = data_1[:,0]
Y = data_1[:,1]
```

```python
# Q3c
print("A log graphical representation of the data")
plt.title("X vs. Log(Y)")
plt.xlabel("X")
plt.ylabel("Log(Y)")
plt.plot(X, np.log(Y), "bo-")
plt.grid(True)
# in the plt.plot line, the "bo-" creates a line with the points marked
# this shows that these points can be any real number
# this implies that is not a discrete set of data
```

```python
# Q3c
print("A log-log graphical representation of the data")
plt.title("Log(X) vs. Log(Y)")
plt.xlabel("Log(X)")
plt.ylabel("Log(Y)")
plt.plot(np.log(X), np.log(Y), "go-")
plt.grid(True)
```
